```python
import pandas as pd
from collections import OrderedDict
import requests
```

```python
#reading DataSet
df=pd.read_csv('Amazon_products_comments.csv')
```

```python
#testing API with 10 records 
df_test=df.loc[0:9]
```

```python
df_test.shape
len(df_test)
```

```python
#importing Boto3 library to call Comprehend API
import boto3 
comprehend = boto3.client('comprehend', region_name='us-east-1')
```

```python
#looping on the test Dataset to pass text to API and print results
for i in range(0,len(df_test)):
      sentiments = comprehend.detect_sentiment(Text=df_test['reviews.text'][i], LanguageCode='en')
      print('--- Sentiment NO. {} ---'.format(i))
      print(sentiments['Sentiment'])
```
